```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set style
sns.set(style="whitegrid")

# 1. Create Sample Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
categories = ["Electronics", "Clothing", "Home Goods"]
np.random.seed(42)

data = {
    "Month": np.tile(months, len(categories)),
    "Category": np.repeat(categories, len(months)),
    "Sales": np.random.randint(100, 500, size=len(months) * len(categories)),
}

df = pd.DataFrame(data)

# 2. Create Multiple Comparative Visualizations
plt.figure(figsize=(16, 10))

# Bar Chart
plt.subplot(2, 2, 1)
sns.barplot(data=df, x="Month", y="Sales", hue="Category")
plt.title("Monthly Sales per Category")

# Line Chart
plt.subplot(2, 2, 2)
sns.lineplot(data=df, x="Month", y="Sales", hue="Category", marker="o")
plt.title("Sales Trends over Time")

# Box Plot
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x="Category", y="Sales")
plt.title("Sales Distribution by Category")

# Heatmap (Correlation example)
# Pivot the data into a matrix for heatmap
pivot = df.pivot_table(index="Category", columns="Month", values="Sales", aggfunc='mean')
plt.subplot(2, 2, 4)
sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".0f")
plt.title("Sales Heatmap")

plt.tight_layout()
plt.show()
```
